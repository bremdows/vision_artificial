## Leer Video

```python
import cv2
import time
cap = cv2.VideoCapture(0)
ftime = 0

while True :
    _, frame = cap.read()
    # Imprime la resolución de la cámara web
    # print(frame.shape)
    ctime = time.time()
    fps = 1/(ctime-ftime)
    ftime = ctime

    # Permite escribir un texto en la imagen mostrada
    """
        1P: Video
        2P: Texto a mostrar
        3P: Ubicacion: (AnchoxAlto)
        4P: Estilo de fuente
        5P: Tamaño de la letra (1.5)
        6P: El color de la fuente en BGR
    """
    cv2.putText(frame, f'fps:{int(ftime)}', (500, 30), cv2.FONT_HERSHEY_PLAIN, 1.5, (255, 0, 0), 2)

    cv2.imshow("Cámara de video", frame)
    if cv2.waitKey(1) & 0XFF==ord("q") :
        break

cv2.destroyAllWindows()
# cap.release()
```
